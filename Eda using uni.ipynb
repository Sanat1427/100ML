```python
import pandas as pd
import seaborn as sns
df= pd.read_csv('train.csv')
df.head()
```

1.Categorical Data
a.Countplot

```python
sns.countplot(df['Embarked'])
```

b.Piechart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

2.Numerical Data

a.Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

b.Displot

```python
sns.displot(df['Age'])
```

c.Boxplot

```python
sns.boxplot(df['Age'])
```
